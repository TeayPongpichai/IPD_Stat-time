```python
#zmmr05
Blanks_AL = zmmr05[(zmmr05['Plant-sp.matl status'].isnull()) & (zmmr05['Ext. Material Group'].isin([1, 2]))] #เก็บข้อมูลที่ Blanks
```

```python
#zmmr05
#Material - Material Description
Normal_Material = Blanks_AL.loc[Blanks_AL['Ext. Material Group'] == 1, 'Material'] #เก็บ Material ของ Normal
Consignment_Material = Blanks_AL.loc[(Blanks_AL['Ext. Material Group'] == 2), 'Material'] #เก็บ Material ของ Consignment
Normal_Material_Description = Blanks_AL.loc[Blanks_AL['Ext. Material Group'] == 1, 'Material Description'] #เก็บ Material Description ของ Consignment
Consignment_Material_Description = Blanks_AL.loc[(Blanks_AL['Ext. Material Group'] == 2), 'Material Description'] #เก็บ Material Description ของ Consignment
```

```python
#zmmr14
# Usage Rate of Normal
total_qty_column_index = zmmr14.columns.get_loc('Total Qty( 6 months )') #เก็บตำแหน่งของคอลัมน์ Total Qty( 6 months )
selected_columns = zmmr14.iloc[:, total_qty_column_index - 3:total_qty_column_index] #เก็บคอลัมน์ 3 คอลัมน์หน้า Total Qty( 6 months )
selected_columns['Material'] = zmmr14['Material'] #เก็บคอลัมน์ 3 คอลัมน์หน้า Total Qty( 6 months ) และมี Material เพื่อบ่งบอกแต่และแถว
```

```python
#zmmr14
# Usage Rate
selected_material_Normal = selected_columns[selected_columns['Material'].isin(Normal_Material)] #เก็บคอลัมน์ 3 คอลัมน์หน้า Total Qty( 6 months ) ของ Normal
selected_material_Consignment = selected_columns[selected_columns['Material'].isin(Consignment_Material)] #เก็บคอลัมน์ 3 คอลัมน์หน้า Total Qty( 6 months ) ของ Consignment

def calculate(row):
    sum_three_months = sum(row.drop('Material'))
    return (sum_three_months * -1) / 90 #สร้างฟังค์ชั่นเพื่อเอาแต่ละแถวมารวมกัน *-1 แล้วหารด้วย 90

selected_material_Normal['Usage Rate'] = selected_material_Normal.apply(calculate, axis=1) #เก็บคอลัมน์ (3เดือนหลัง),Material,Usage Rate  ของ Normal
selected_material_Consignment['Usage Rate'] = selected_material_Consignment.apply(calculate, axis=1) #เก็บคอลัมน์ (3เดือนหลัง),Material,Usage Rate  ของ Consignment

Usage_material_Normal = selected_material_Normal[['Material', 'Usage Rate']] #เก็บคอลัมน์ Material,Usage Rate ของ Normal
Usage_material_Consignment = selected_material_Consignment[['Material', 'Usage Rate']] #เก็บคอลัมน์ Material,Usage Rate Consignment
```

```python
#Material, Material Description------------------

Normal = pd.DataFrame({
    'Material': Normal_Material,
    'Material Description': Normal_Material_Description
}) #นำข้อมูลที่สร้างขึ้นใหม่มาใส่ในคอลัมน์ Material,Material Description ของ Nomal

Consignment = pd.DataFrame({
    'Material': Consignment_Material,
    'Material Description': Consignment_Material_Description
}) #นำข้อมูลที่สร้างขึ้นใหม่มาใส่ในคอลัมน์ Material,Material Description ของ Consignment

#Material, Material Description------------------
```

```python
# Usage Rate------------------

# เลือกข้อมูล Usage Rate ที่ Material ตรงกัน
# Merge 'Usage Rate' column from Usage_material_Normal into Normal DataFrame
Normal = Normal.merge(Usage_material_Normal[['Material', 'Usage Rate']], on='Material', how='left')

# Merge 'Usage Rate' column from Usage_material_Consignment into Consignment DataFrame
Consignment = Consignment.merge(Usage_material_Consignment[['Material', 'Usage Rate']], on='Material', how='left')

# แทนที่ค่า NaN ใน Material ด้วยค่าก่อนหน้า
Normal['Material'].fillna(method='bfill', inplace=True)
Consignment['Material'].fillna(method='bfill', inplace=True)

# เรียงลำดับ Normal และ Consignment ตามคอลัมน์ Usage Rate จากมากไปน้อย
Normal = Normal.sort_values(by='Usage Rate', ascending=False)
Consignment = Consignment.sort_values(by='Usage Rate', ascending=False)

# แปลงค่า Usage Rate ที่เป็น 0.00 หรือ NaN เป็น No usage และจัดรูปแบบทศนิยมสองตำแหน่ง
Normal['Usage Rate'] = Normal['Usage Rate'].apply(lambda x: 'No usage' if np.isnan(x) or abs(x - 0.00) < 0.01 else f'{x:.2f}')
Consignment['Usage Rate'] = Consignment['Usage Rate'].apply(lambda x: 'No usage' if np.isnan(x) or abs(x - 0.00) < 0.01 else f'{x:.2f}')

# Usage Rate------------------
```

```python
#'Min', 'Max'-------
Quantity = mb51.groupby(['Material', 'Posting Date'])['Quantity'].agg(lambda x: sum(x * -1)).reset_index()

min_max_values = Quantity.groupby('Material')['Quantity'].agg(['min', 'max']).reset_index()

min_max_values.columns = ['Material', 'Min', 'Max']

Normal = pd.merge(Normal, min_max_values[['Material', 'Min', 'Max']], on='Material', how='left')
Consignment = pd.merge(Consignment, min_max_values[['Material', 'Min', 'Max']], on='Material', how='left')
#'Min', 'Max'-------
```
